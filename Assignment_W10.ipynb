```python
import mesa
import numpy as np
```

```python
totalsteps=10
```

```python
class Patient(mesa.Agent):
    
    # initialization
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.state = "outside"
        self.model = model
        self.appTime=np.random.randint(totalsteps)
        self.severity=np.random.randint(1,11)
        print("Initializing Patient", self.unique_id, self.state,self.severity, self.appTime)
        
    def step(self):
        print("Patient Before", self.unique_id, self.state,self.severity, self.appTime)
        if self.state == 'outside':
            if self.appTime == self.model.schedule2.time and (self.severity==9 or self.severity==10):
                self.state='hospital'
                print('agent arrived', self.unique_id)
            elif self.appTime == self.model.schedule2.time and self.severity<9 :
                self.state='arrived'
        elif self.state == 'arrived':
            for t in self.model.schedule1.agents:
                if t.state=='free':
                    self.state='triage'
                    t.state='busy'
                    t.patient=self
                    t.triagestart=self.model.schedule2.time
                    t.severity=self.severity
                    break
      
        
    
```

```python
class Triage(mesa.Agent):
    
    # initialization
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.state = "free"
        self.model = model
        self.patient = None
        self.triagestart=0
        self.severity=0
        print("Initializing triage agent", self.unique_id, self.state)
        
    def step(self):
        print("Triage", self.unique_id, self.state,self.patient)
        if self.state=='busy':
            if self.model.schedule2.time>=self.triagestart+self.severity:
                self.patient.state='after triage'
                self.state='free'
    
```

```python
class EDModel(mesa.Model):
    def __init__(self, N, M,):
        self.num_patients = N
        self.num_triage = M
        self.schedule1 = mesa.time.BaseScheduler(self)
        self.schedule2 = mesa.time.BaseScheduler(self)
        
        for i in range(self.num_triage):
            a = Triage(i, self)
            self.schedule1.add(a)
        
        for i in range(self.num_patients):
            a = Patient(i, self)
            self.schedule2.add(a)
     
    # reqiured function that completes one step of simulation
    def step(self):
        # run clinicians
        self.schedule1.step()
        
        # run patients
        self.schedule2.step()
```
